### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

#### Note: Also add title, xlabel, and ylabel to the plot.

#### A scatter plot is a type of plot used to visualize the relationship between two numerical variables. In a scatter plot, each data point is represented as a single point on the plot, with its position determined by the values of the two variables. One variable is typically plotted along the x-axis (horizontal axis), while the other variable is plotted along the y-axis (vertical axis).

```python
import matplotlib.pyplot as plt
plt.scatter(x,y, alpha=.5)
```

### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
#### Use the following data:

```python
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
```

#### The subplot() function in Matplotlib is used to create multiple subplots within the same figure. It allows you to divide the figure into a grid of rows and columns and specify which subplot to plot in each cell of the grid.

```python
import matplotlib.pyplot as plt

# Generate some sample data
x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplots
plt.figure(figsize=(10, 8))

# Subplot 1
plt.subplot(4, 1, 1)
plt.plot(x, y1)
plt.title('Subplot 1')

# Subplot 2
plt.subplot(4, 1, 2)
plt.plot(x, y2)
plt.title('Subplot 2')

# Subplot 3
plt.subplot(4, 1, 3)
plt.plot(x, y3)
plt.title('Subplot 3')

# Subplot 4
plt.subplot(4, 1, 4)
plt.plot(x, y4)
plt.title('Subplot 4')

# Adjust layout
plt.tight_layout()

# Show plots
plt.show()
```

### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

#### A bar plot is a type of plot in which rectangular bars are used to represent the frequencies or counts of categories or numerical values. The length or height of each bar corresponds to the value it represents. Bar plots are commonly used to visualize categorical data or to compare numerical values across different categories.

```python
plt.bar(company, profit)
plt.title("Companies with profits")
plt.xlabel("Company")
plt.ylabel("Profits")
```
